```python


def removelabel(graph):
    len_ver=len(graph.keys())
    ver=list(graph.keys())
    for i in range(len_ver):
        try:
            for j in range(len(graph[f'{ver[i]}'])):
                graph[f'{ver[i]}'][j]=graph[f'{ver[i]}'][j][1]
        except:
            pass
    return graph

def connection(graph):
    len_ver=len(graph.keys())
    ver=list(graph.keys())
    list_of_connection=[]
    keys=graph.keys()
    list_of_vertices=[list(keys)[i].strip('][').split(', ') for i in range(len(graph))]
    for i in range(len(list_of_vertices)):
        for j in range(len(list_of_vertices[0])):
            list_of_vertices[i][j]=int(list_of_vertices[i][j])
            
    for i in range(len_ver):
        for j in range(len_ver):
            if list_of_vertices[j] in graph[f'{ver[i]}']:
                list_of_connection.append([i,j])
    return [list_of_vertices,list_of_connection]
    
graph_pure=removelabel(graph)
lists=connection(graph_pure)
```
